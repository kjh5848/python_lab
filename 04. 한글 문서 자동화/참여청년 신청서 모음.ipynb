```python
from utill import excel_utill, hwp_utill
import os

# 파일 경로 및 시트 설정
EXCEL_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\python_lab\07. 한글 문서\한글 자동화_엑셀파일.xlsm"
EXCEL_SHEET_NAME = "참여자명단"
HWP_FILE_NAME = "참여청년 신청서 모음"
HWP_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\python_lab\07. 한글 문서\★누름틀_문서\2. 참여청년 서류"
SAVE_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\참여신청서 모음"

# 객체 생성
EObject = excel_utill.ExcelUtill()
HObject = hwp_utill.HwpUtill()

# 엑셀 데이터 불러오기
excel_data = EObject.엑셀열기(EXCEL_FILE_PATH, EXCEL_SHEET_NAME, header=1)

# 메인 작업 함수
def 한글문서_생성(excel_data, hwp_file_name, hwp_file_path):
    for index, row in excel_data.iterrows():
        
        if not row['성명']:
            continue

        try:
            # 파일 열기
            HObject.파일열기(hwp_file_name, hwp_file_path)

            # 필드 채우기
            HObject.채우기('성명', row['성명'])
            HObject.채우기('주민등록번호', row['주민등록번호'])
            HObject.채우기('연락처', row['연락처'])
            HObject.채우기('이메일', row['이메일'])
            HObject.채우기('최종학과', row['최종학과'])
            HObject.채우기('작성날짜', row['작성날짜'])
            HObject.채우기('최종학교명', row['최종학교명'])
            HObject.채우기('생년월일', row['생년월일'])
            HObject.채우기('주소', row['주소'])
            HObject.채우기('만나이', row['만나이'])
            

            if row['직무'] == 'IT':
                HObject.채우기('IT', '■')
            elif row['직무'] == '광고 마케팅':
                HObject.채우기('광고 마케팅', '■')


            # 최종학력구분에 따른 처리
            if row['최종학력구분'] == '대학교_4년제':
                HObject.채우기('대학교', '■')
                HObject.채우기('대학교명', row['최종학교명'])

                if row['최종학력상태'] == '졸업':
                    HObject.채우기('대학교졸업', '■')
                elif row['최종학력상태'] == '졸업예정자':
                    HObject.채우기('졸업예정자', '■')
                elif row['최종학력상태'] in ['휴학', '재학']:
                    if row['학년'] in ['1', '2']:
                        HObject.채우기('재학(1~2학년)', '■')
                    elif row['학년'] in ['3', '4']:
                        HObject.채우기('재학(3~4학년)', '■')

            elif row['최종학력구분'] in ['대학교_2년제', '대학교_2/3년제']:
                HObject.채우기('대학교', '■')
                HObject.채우기('대학교명', row['최종학교명'])
                
                if row['최종학력상태'] == '졸업':
                    HObject.채우기('대학교졸업', '■')
                elif row['최종학력상태'] == '졸업예정자':
                    HObject.채우기('졸업예정자', '■')
                elif row['최종학력상태'] in ['휴학', '재학']:
                    if row['학년'] in ['1', '2']:
                        HObject.채우기('재학(1~2학년)', '■')
                    elif row['학년'] in ['3']:
                        HObject.채우기('재학(3~4학년)', '■')

            elif row['최종학력구분'] == '고등학교':
                HObject.채우기('고등학교', '■')
                HObject.채우기('고등학교명', row['최종학교명'])

                if row['최종학력상태'] == '졸업':
                    HObject.채우기('고등학교졸업', '■')
                elif row['최종학력상태'] == '졸업예정자':
                    HObject.채우기('고등학교졸업예정자', '■')
                elif row['최종학력상태'] == '재학':
                    if row['학년'] in ['1', '2']:
                        HObject.채우기('고등학교재학(1~2학년)', '■')
                    elif row['학년'] == '3':
                        HObject.채우기('고등학교재학(3학년)', '■')
                elif row['최종학력상태'] == '검정고시':
                    HObject.채우기('검정고시', '■')     
                    
            # 폴더 생성
            if not os.path.exists(SAVE_FILE_PATH):
                os.makedirs(SAVE_FILE_PATH)

            # 파일 저장
            new_file_name = f"{hwp_file_name}_{row['성명']}_{row['생년월일']}.hwp"
            HObject.저장(new_file_name, SAVE_FILE_PATH)
    
            print(f"파일 저장 완료: {new_file_name}")
            HObject.닫기()
            
        except Exception as e:
            print(f"오류 발생 (행 {index + 1}): {e}")

    # 작업 완료 후 한글 종료
    HObject.종료()

# 실행
한글문서_생성(excel_data, HWP_FILE_NAME, HWP_FILE_PATH)
```
