```python
import os
from utill import hwp_utill

def 한글문서_이미지삽입(HObject, hwp_file_name, hwp_file_path, save_file_path, image_dir, 성명):
    """한글 문서에 이미지를 삽입하는 샘플 코드"""
    try:
        # 한글 파일 열기
        HObject.파일열기(hwp_file_name, hwp_file_path)
        
        # 한글 파일에서 필드 이름 가져오기
        field_names = HObject.필드가져오기(hwp_file_name, hwp_file_path)
        
        
        def insert_signature_image(HObject, image_path, target_field_name, width=19, height=19):
            """지정된 필드에 서명 이미지 삽입 및 크기/위치 설정"""
            try:
                HObject.hwp.MoveToField(target_field_name)
                HObject.hwp.InsertPicture(image_path, Embedded=True, sizeoption=0)
                HObject.hwp.FindCtrl()
                HObject.hwp.HAction.GetDefault("ShapeObjDialog", HObject.hwp.HParameterSet.HShapeObject.HSet)

                # 이미지 설정 적용
                HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("TreatAsChar", 0)
                HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("TextWrap", 3)
                # 개체의 기준을 "문단"으로 설정
                # HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("HorzRelTo", 0)  # 문단 기준 정렬
                # HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("VertRelTo", 0)  # 문단 기준 정렬
                HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("Width", int(width * 250.465))
                HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("Height", int(height * 80.465))
                HObject.hwp.HParameterSet.HShapeObject.HSet.SetItem("HorzOffset", HObject.hwp.MiliToHwpUnit(6.5))
             
                HObject.hwp.HAction.Execute("ShapeObjDialog", HObject.hwp.HParameterSet.HShapeObject.HSet)

                print(f"서명 이미지 삽입 완료: {target_field_name}")
            
            except Exception as e:
                print(f"이미지 삽입 오류 ({target_field_name}): {e}")

        
        def 필드에_이미지_삽입(HObject, field_names, image_dir, name):
            """필드명이 {서명}이고, 성명과 같은 이미지 파일이 있으면 삽입"""
            valid_extensions = [".jpg", ".png", ".jpeg"]
            signature_fields = ["서명", "서명1", "서명2"]

            for field_name in signature_fields:
                if field_name in field_names:
                    image_path = None
                    for ext in valid_extensions:
                        temp_path = os.path.join(image_dir, f"{name}{ext}")
                        if os.path.exists(temp_path):
                            image_path = temp_path
                            break
                    if image_path:
                        insert_signature_image(HObject, image_path, field_name)
                else:
                    print(f"이미지 없음: {name}")
        
        # 이미지 삽입 실행
        필드에_이미지_삽입(HObject, field_names, image_dir, 성명)

        # 저장
        new_file_name = f"{hwp_file_name}_{성명}.hwp"
        HObject.저장(new_file_name, save_file_path)
        print(f"파일 저장 완료: {new_file_name}")
    
    except Exception as e:
        print(f"오류 발생 (성명: {성명}): {e}")
    
    # finally:
        # # 한글 파일 닫기
        # HObject.닫기()

# 저장된 파일 폴더 열기
    # HObject.저장파일열기(save_file_path)

    # 한글 종료
    # HObject.종료()
# 객체 생성
HObject = hwp_utill.HwpUtill()

# 샘플 실행
한글문서_이미지삽입(
    HObject=HObject,
    hwp_file_name="참여청년 신청서 모음",
    hwp_file_path=r"C:\Users\Administrator\Desktop\미래내일일경험 공유\python_lab\07. 한글 문서\★누름틀_문서\2. 참여청년 서류",
    save_file_path=r"C:\Users\Administrator\Desktop\미래내일일경험 공유",
    image_dir=r"C:\Users\Administrator\Desktop\서명\쇼우테크 11차",
    성명="김선영"
)
```
