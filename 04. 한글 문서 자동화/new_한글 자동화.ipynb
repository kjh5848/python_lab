```python
from utill import excel_utill, hwp_utill
import pandas as pd
import os


# 파일 경로 및 시트 설정
EXCEL_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\python_lab\07. 한글 문서\new_한글 자동화.xlsx"
EXCEL_SHEET_NAME = "참여자명단"
HWP_FILE_NAME = "참여청년 신청서 모음"
HWP_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\python_lab\07. 한글 문서\★누름틀_문서\2. 참여청년 서류"
SAVE_FILE_PATH = r"C:\Users\Administrator\Desktop\미래내일일경험 공유\신청서 모음2"


# 객체 생성
EObject = excel_utill.ExcelUtill()
HObject = hwp_utill.HwpUtill()



# 엑셀 데이터 불러오기
excel_data = EObject.엑셀열기(EXCEL_FILE_PATH, EXCEL_SHEET_NAME, header=1)



def get_hwp_field_names(hwp_object, file_name, file_path):
    """HWP 파일에서 필드 목록 가져오기"""
    hwp_object.파일열기(file_name=file_name, file_path=file_path)
    field_list = hwp_object.hwp.GetFieldList().split('\x02')
    return [field.split('\x00')[0] for field in field_list]


def fill_hwp_fields(hwp_object, field_names, row_data):
    """HWP 필드에 데이터를 채우기"""
    for field_name in field_names:
        if field_name in row_data.index:
            cell_data = row_data[field_name]
            if pd.isna(cell_data):
                field_text = ""
            elif isinstance(cell_data, (int, float)):
                # 숫자 데이터 포맷팅
                field_text = f"{int(cell_data)}" if cell_data % 1 == 0 else f"{cell_data:.2f}"
            else:
                field_text = str(cell_data)
            HObject.채우기(field_name, field_text)
            HObject.학력정보입력(row=row_data)
            if row_data['직무'] == 'IT':
                HObject.채우기('IT', '■')
            elif row_data['직무'] == '광고 마케팅':
                HObject.채우기('광고 마케팅', '■')


def save_hwp_file(hwp_object, save_path, file_name, row_data):
    """HWP 파일 저장"""
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    new_file_name = f"{HWP_FILE_NAME}_{row_data['성명']}_{row_data['생년월일']}.hwp"
    hwp_object.저장(new_file_name, save_path)
    print(f"파일 저장 완료: {new_file_name}")
    hwp_object.닫기()


def main():
    # 엑셀 데이터 로드
    excel_data = load_excel_data(EXCEL_FILE_PATH, EXCEL_SHEET_NAME)

    # HWP 필드 이름 가져오기

    # 각 행에 대해 HWP 작업 수행
    for idx, row in excel_data.iterrows():
        field_names = get_hwp_field_names(HObject, HWP_FILE_NAME, HWP_FILE_PATH)
        fill_hwp_fields(HObject, field_names, row)
        save_hwp_file(HObject, SAVE_FILE_PATH, HWP_FILE_NAME, row)

    # HWP 종료
    HObject.종료()

    os.startfile(SAVE_FILE_PATH)


if __name__ == "__main__":
    main()
```
