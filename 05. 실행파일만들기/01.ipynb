```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QVBoxLayout, QLabel, QMessageBox

class MyToolbox(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('PyQt5 도구 상자')
        self.setGeometry(100, 100, 300, 200)

        # 레이아웃 설정
        layout = QVBoxLayout()

        # 라벨 추가
        self.label = QLabel('기능을 선택하세요:', self)
        layout.addWidget(self.label)

        # 버튼 설정
        self.button1 = QPushButton('작업 1', self)
        self.button1.clicked.connect(self.on_click_task1)
        layout.addWidget(self.button1)

        self.button2 = QPushButton('작업 2', self)
        self.button2.clicked.connect(self.on_click_task2)
        layout.addWidget(self.button2)

        self.setLayout(layout)

    def on_click_task1(self):
        QMessageBox.information(self, '작업 1', '작업 1이 수행되었습니다.')

    def on_click_task2(self):
        QMessageBox.information(self, '작업 2', '작업 2가 수행되었습니다.')

# PyQt5 애플리케이션 실행
if __name__ == '__main__':
    app = QApplication(sys.argv)
    toolbox = MyToolbox()
    toolbox.show()
    sys.exit(app.exec_())
```
